### Styling the graphs

```python
x = [1, 2, 3, 4, 5, 6]
y = [1 ,4, 9, 16, 25, 36]
plt.plot(x, y, 'r*')
plt.show()
```

```python
x = np.arange(1, 12)
y = x**2
plt.title('Sqaure values')
plt.xlabel('numbers')
plt.ylabel('squares')
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(-10, 150)
plt.show()
```

```python
x1 = np.arange(-10, 20)
y1 = x1**2
x2 = np.arange(-10, 20)
y2 = x2 + 50
plt.plot(x1, y1, label = 'Quadratic')
plt.plot(x2, y2, label = 'Linear')
plt.text(7, 200, 'Hello')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x1, y1, 'b', x2, y2, 'ro')
plt.show()
```

### Bar Graph

```python
plt.bar([1, 2, 3], [10, 20, 5], width = 0.2, color = 'b', label='discount')
plt.bar([1.5, 2.5, 3.5], [5, 10, 15], width = .2, color = 'r', label = 'cost')
plt.legend()
plt.show()
```

### Histogram

```python
vals = np.random.rand(100)
plt.hist(vals, [0, 0.5, 0.7, 0.8, 0.9, 1])
plt.show()
```

### Normal Distributions

```python
mu = 100
sigma = 15
s = np.random.normal(mu, sigma, 1000)
```

```python
count, bins, patches = plt.hist(s, 20)
plt.show()
```

### Scatter Plots

```python
x1 = np.random.rand(20)
y1 = np.random.rand(20)

x2 = np.random.rand(20)
y2 = np.random.rand(20)

np.random.seed(42)

plt.scatter(x1, y1)
plt.scatter(x2, y2, color = 'r')
plt.show()
```

### Pie Charts

```python
plt.pie([10, 20 ,30, 40],
       labels = ['Mech', 'Elec.', 'ECE', 'CSE'],
       colors = ['c', 'm', 'r', 'b'],
       shadow = True,
       startangle = 90,
       explode = (0,.2,0,0))
plt.show()
```

##### Sigmoid Function

```python
sigmoid = lambda x:1/(1 + math.exp(-x))
x = np.arange(-10, 10)
y = [sigmoid(i) for i in x]
plt.plot(x, y)
plt.show()
```
